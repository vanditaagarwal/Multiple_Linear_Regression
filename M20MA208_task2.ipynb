```python
# Handling Missing Entries denoted with '?'

df['normalized-losses'].replace('?',0 ,inplace=True)
df['normalized-losses']=df['normalized-losses'].astype(int)
df['normalized-losses'].replace(0,df['normalized-losses'].mean(),inplace=True)
df

df['num-of-doors'].replace('?',np.nan,inplace=True)

df['bore'].replace('?',0 ,inplace=True)
df['bore']=df['bore'].astype(float)
df['bore'].replace(0,df['bore'].mean(),inplace=True)


df['stroke'].replace('?',0 ,inplace=True)
df['stroke']=df['stroke'].astype(float)
df['stroke'].replace(0,df['stroke'].mean(),inplace=True)


df['horsepower'].replace('?',0 ,inplace=True)
df['horsepower']=df['horsepower'].astype(int)
df['horsepower'].replace(0,df['horsepower'].mean(),inplace=True)


df['peak-rpm'].replace('?',0 ,inplace=True)
df['peak-rpm']=df['peak-rpm'].astype(int)
df['peak-rpm'].replace(0,df['peak-rpm'].mean(),inplace=True)


df['price'].replace('?',0 ,inplace=True)
df['price']=df['price'].astype(int)
df['price'].replace(0,df['price'].mean(),inplace=True)
df
```

```python
# Dropping Null Values

df_1=df.dropna()
df_1
```

```python
# Assigning dummy values to categorical columns

df_2=pd.get_dummies(df_1,columns=['make','fuel-type','aspiration','num-of-doors','body-style','drive-wheels',
                                'engine-location','engine-type','num-of-cylinders','fuel-system'], drop_first=True)
df_2
```

```python
# Normalizing featues data 'x'
x = (x - np.min(x))/(np.max(x)-np.min(x)).values
```

```python
# Train-Test Split (60:40)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.40,random_state=0)
```

```python
# Defining Linear Regression on Multiple Features

class LinearRegression() :

    def __init__( self, lr, itr ) :
        self.learning_rate = lr
        self.iterations = itr

    def fit( self, X, Y ) :
        samples, features = X.shape
        self.W = np.zeros(features)
        self.b = 0
        self.X = X.values
        self.Y = Y

        # optimization using gradient descent
        for i in range( self.iterations ) :
            Y_pred = np.dot(self.X, self.W) + self.b

            # calculating gradients
            dW = - (2*(self.X.T).dot(self.Y - Y_pred))/samples
            db = - 2 * np.sum( self.Y - Y_pred )/samples

            # updating weights
            self.W = self.W - self.learning_rate * dW
            self.b = self.b - self.learning_rate * db

    def predict( self, X ):
        answer = X.dot(self.W)+self.b
        return answer
```

```python
# Implementing Linear Regression on training data

multilinreg = LinearRegression( lr = 0.001,itr = 1000)
multilinreg.fit( x_train, y_train )
```

```python
# Predicting values for test data

y_pred = multilinreg.predict(x_test)
#y_pred1 = multilinreg.predict(x_train)

print(y_pred)
print(y_test)

#print( "Predicted values ", np.round( y_pred.values[:10], 2 ) )
#print( "Real values      ", y_test[:10] )
```

```python
# Mean Square Error

from sklearn.metrics import mean_squared_error
Mse = mean_squared_error(y_test, y_pred,squared =False)
print("MSE", Mse)
```
